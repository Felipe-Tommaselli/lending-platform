```python
# Treinamento do modelo
model = RandomForestClassifier()
model.fit(X, y)
```

```python
# Função para prever risco
def predict_risk(credit_score, balance, debt):
    prediction = model.predict([[credit_score, balance, debt]])
    return True if prediction[0] == 1 else False
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from mpl_toolkits.mplot3d import Axes3D

# Carregar os dados
with open("loan_data.json", "r") as f:
    data = json.load(f)

X = np.array(data["X"])
y = np.array(data["y"])

# Treinar o modelo de Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X, y)

# Definir o intervalo para cada variável
credito_range = np.linspace(X[:, 0].min() - 100, X[:, 0].max() + 100, 50)
saldo_range = np.linspace(X[:, 1].min() - 100, X[:, 1].max() + 100, 50)
dividas_range = np.linspace(X[:, 2].min() - 100, X[:, 2].max() + 100, 50)

# Criar uma grade com os três eixos
xx, yy, zz = np.meshgrid(credito_range, saldo_range, dividas_range)

# Prever com o modelo em toda a grade
Z = model.predict(np.c_[xx.ravel(), yy.ravel(), zz.ravel()])
Z = Z.reshape(xx.shape)

# Plotar em 3D
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Usar scatter para exibir as previsões em 3D
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y, cmap=plt.cm.RdYlBu, s=40, edgecolor='k')

# Configurar eixos e rótulos
ax.set_xlabel('Crédito Disponível')
ax.set_ylabel('Saldo em Conta')
ax.set_zlabel('Dívidas')
ax.set_title('Fronteiras de Decisão em 3D - Random Forest')

plt.show()
```
